```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib_inline

# Function to compute the number of subcoalitions with n customers
def compute_num_subcoalitions(n):
    return 2**n - 1

# Function to create and save plot with number of subcoalitions up to max_n
def fun_create_plot(max_n, view_dict=True, language='english'):

    # Save number of subcoalitions up to max_n in dictionary
    subcoalitions_dict = {}
    for i in range(max_n+1):
        subcoalitions_dict['n=' + str(i)] = compute_num_subcoalitions(i)
    if (view_dict == True): display(subcoalitions_dict)

    # Create plot
    plt.figure(figsize=(10, 4))
    plt.plot(subcoalitions_dict.values(), color='b')
    plt.xticks(np.arange(start=0, stop=max_n+1, step=1))

    # Make english or german labels
    if (language == 'english'):
        plt.xlabel('Number of customers n', size=12)
        plt.ylabel('Number of possible subcoalitions', size=12)
        plt.title('Exponential increase in possible subcoalitions with increasing n', size=16)
        name = ('shapley_value_exponential_number_of_subcoalitions.png')
    elif (language == 'german'):
        plt.xlabel('Anzahl der Kunden n', size=12)
        plt.ylabel('Anzahl aller möglichen Subkoalitionen', size=12)
        plt.title('Exponentieller Anstieg der möglichen Subkoalitionen bei steigendem n', size=16)
        name = ('shapley_value_exponentielle_anzahl_subkoalitionen.png')
    plt.grid()
    plt.savefig(name)
    plt.show()
    return subcoalitions_dict

# Execute function to create and save plot
subcoalitions_dict = fun_create_plot(max_n=15, view_dict=False, language='german')
```
