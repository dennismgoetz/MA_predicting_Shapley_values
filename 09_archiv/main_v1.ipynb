```python
import json
from IPython.utils.capture import capture_output

# Define settings
optimization_problem = "CVRP"

# Save the settings in a file
with open("settings.json", "w") as file:
    json.dump({"optimization_problem": optimization_problem}, file)

###############################################################################
# TSP
###############################################################################
if (optimization_problem == "TSP"):
    # Run all scripts to store the best parameters
    with capture_output() as captured:
        %run a1_Linear.ipynb
    print("Script 'a1_Linear.ipynb' completed!")

    with capture_output() as captured:
        %run a2_Tree.ipynb
    print("Script 'a2_Tree.ipynb' completed!")

    with capture_output() as captured:
        %run a3_Ensembles.ipynb
    print("Script 'a3_Ensembles.ipynb' completed!")

    with capture_output() as captured:
        %run a4_SVM_pipe.ipynb
    print("Script 'a4_SVM_pipe.ipynb' completed!")

    with capture_output() as captured:
        %run a5_NN.ipynb
    print("Script 'a5_NN.ipynb' completed!")

    # Run the final script to store the test score results of the tuned models
    with capture_output() as captured:
        %run a6_all_models.ipynb
    print("Script 'a6_all_models.ipynb' completed!")

###############################################################################
# CVRP
###############################################################################
elif (optimization_problem == "CVRP"):
    # Run all scripts to store the best parameters
    with capture_output() as captured:
        %run a1_Linear.ipynb
    print("Script 'a1_Linear.ipynb' completed!")

    with capture_output() as captured:
        %run a3_Ensembles.ipynb
    print("Script 'a3_Ensembles.ipynb' completed!")

    with capture_output() as captured:
        %run a5_NN.ipynb
    print("Script 'a5_NN.ipynb' completed!")

    # Run the final script to store the test score results of the tuned models
    with capture_output() as captured:
        %run a6_all_models.ipynb
    print("Script 'a6_all_models.ipynb' completed!")
```
